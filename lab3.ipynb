```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

operations = pd.read_csv('operations.csv', sep=',', decimal=',')
dataset = pd.read_csv('2 - dataset.csv', sep=',', decimal=',')
types = pd.read_csv('3 - types.csv', sep=';', decimal=',')
codes = pd.read_csv('4 - codes.csv',sep=';')
newset = pd.merge(operations, types, on = 'type')
sets = pd.merge(newset, codes, on = 'code')
finalmerge= pd.merge(sets,dataset,on='client_id')
finalmerge[:1000]

```

```python
finalmerge['client_id'].describe()
```

```python
import matplotlib.pyplot as plt
finalmerge[finalmerge['client_id']>51500000][finalmerge['client_id']<91500000]['client_id'].hist(bins=50,color = 'green')
plt.xlabel("Client_id",fontsize=15)
plt.ylabel("Transaction",fontsize=15)
plt.show()
#transactions by client_id
```
